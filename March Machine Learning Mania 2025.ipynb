```python
def create_team_stats(df):
    """
    Convert Detailed Results into a long format of (Season, TeamID)'s stats:
    - PF, PA, FGM, FGA, etc. from the perspective of that TeamID
    """
    # Winners
    w_cols = [
        "Season","WTeamID","WScore","LScore",
        "WFGM","WFGA","WFGM3","WFGA3","WFTM","WFTA",
        "WOR","WDR","WAst","WTO","WStl","WBlk","WPF"
    ]
    winners = df[w_cols].copy()
    winners.rename(columns={
        "WTeamID":"TeamID","WScore":"FScore","LScore":'AScore',
        "WFGM":"FGM","WFGA":"FGA","WFGM3":"FGM3","WFGA3":"FGA3",
        "WFTM":"FTM","WFTA":"FTA","WOR":"OR","WDR":"DR","WAst":"Ast",
        "WTO":"TO","WStl":"Stl","WBlk":"Blk","WPF":"PFoul"
    }, inplace=True)
    winners["Result"] = 1

    # Losers
    l_cols = [
        "Season","LTeamID","WScore","LScore",
        "LFGM","LFGA","LFGM3","LFGA3","LFTM","LFTA",
        "LOR","LDR","LAst","LTO","LStl","LBlk","LPF"
    ]
    losers = df[l_cols].copy()
    losers.rename(columns={
        "LTeamID":"TeamID","LScore":"FScore","WScore":'AScore',
        "LFGM":"FGM","LFGA":"FGA","LFGM3":"FGM3","LFGA3":"FGA3",
        "LFTM":"FTM","LFTA":"FTA","LOR":"OR","LDR":"DR","LAst":"Ast",
        "LTO":"TO","LStl":"Stl","LBlk":"Blk","LPF":"PFoul"
    }, inplace=True)
    losers["Result"] = 0

    return pd.concat([winners, losers], axis=0, ignore_index=True)
```

```python


def add_binary_rows(df):
    """
    For a Detailed Tourney DF with WTeamID, LTeamID, etc.,
    create two rows per game:
      1) TeamA=winner, label=1
      2) TeamA=loser, label=0
    """
    rename_winner = {
        "WTeamID":"TeamA","WScore":"ScoreA",'W_Seed_Rank': 'A_Seed_Rank', 'W_Team_Region':'A_Team_Region', 'W_Exp':'A_Exp', 'W_T_Region':'A_T_Region',
        "WFGM":"AFGM","WFGA":"AFGA","WFGM3":"AFGM3","WFGA3":"AFGA3","WFTM":"AFTM","WFTA":"AFTA","WOR":"AOR","WDR":"ADR","WAst":"AAst","WTO":"ATO","WStl":"AStl","WBlk":"ABlk","WPF":"APF",
        "LTeamID":"TeamB","LScore":"ScoreB", 'L_Seed_Rank': 'B_Seed_Rank', 'L_Team_Region':'B_Team_Region', 'L_Exp':'B_Exp', 'L_T_Region':'B_T_Region',
        "LFGM":"BFGM","LFGA":"BFGA","LFGM3":"BFGM3","LFGA3":"BFGA3","LFTM":"BFTM","LFTA":"BFTA","LOR":"BOR","LDR":"BDR","LAst":"BAst","LTO":"BTO","LStl":"BStl","LBlk":"BBlk","LPF":"BPF"
    }
    rename_loser = {
        "WTeamID":"TeamB","WScore":"ScoreB", 'W_Seed_Rank': 'B_Seed_Rank', 'W_Team_Region':'B_Team_Region', 'W_Exp':'B_Exp', 'W_T_Region':'B_T_Region',
        "WFGM":"BFGM","WFGA":"BFGA","WFGM3":"BFGM3","WFGA3":"BFGA3","WFTM":"BFTM","WFTA":"BFTA","WOR":"BOR","WDR":"BDR","WAst":"BAst","WTO":"BTO","WStl":"BStl","WBlk":"BBlk","WPF":"BPF",
        "LTeamID":"TeamA","LScore":"ScoreA", 'L_Seed_Rank': 'A_Seed_Rank', 'L_Team_Region':'A_Team_Region', 'L_Exp':'A_Exp', 'L_T_Region':'A_T_Region',
        "LFGM":"AFGM","LFGA":"AFGA","LFGM3":"AFGM3","LFGA3":"AFGA3","LFTM":"AFTM","LFTA":"AFTA","LOR":"AOR","LDR":"ADR","LAst":"AAst","LTO":"ATO","LStl":"AStl","LBlk":"ABlk","LPF":"APF"
    }

    df_win = df.rename(columns=rename_winner).copy()
    df_lose = df.rename(columns=rename_loser).copy()

    df_win["Label"] = 1
    df_lose["Label"] = 0

    return pd.concat([df_win, df_lose], ignore_index=True)

df_games_mod = add_binary_rows(df_games)
```

```python
# Create function to remove null values
def replace_Na_with_mean_a(df):
    lst = stats_agg_3.columns.tolist()
    lst.remove('TeamID')
    for c in lst:
        c_1 = 'A'+'_'+c
        df[c_1] = np.where(df[c_1].isnull(), df[c],df[c_1])
        df[c_1].fillna(0, inplace = True)
        df.drop(c, axis = 1, inplace = True)
def replace_Na_with_mean_b(df):
    lst = stats_agg_3.columns.tolist()
    lst.remove('TeamID')
    for c in lst:
        c_1 = 'B'+'_'+c
        df[c_1] = np.where(df[c_1].isnull(), df[c],df[c_1])
        df[c_1].fillna(0, inplace = True)
        df.drop(c, axis = 1, inplace = True)
# fill with zero for a team vs b team data is not available
def fill_na(df):
    for c in df.columns:
        if c[:3]== 'AB_':
            df[c].fillna(0, inplace = True)
        elif c[:3]== 'Exp':
            df[c].fillna(0, inplace = True)
```
